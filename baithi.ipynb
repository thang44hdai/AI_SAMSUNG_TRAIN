```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Đọc dữ liệu từ file CSV
data = pd.read_csv('Train_Samsung_imputed.csv')

# Chọn các cột cần chuyển đổi
columns_to_encode = ['X1', 'X2', 'X3', 'X4', 'X5', 'X11']
data_to_encode = data[columns_to_encode]

# Khởi tạo bộ mã hóa OneHot
encoder = OneHotEncoder()
encoder.fit(data_to_encode)

# Chuyển đổi dữ liệu
encoded_data = encoder.transform(data_to_encode).toarray()

# Thêm dữ liệu đã chuyển đổi vào DataFrame ban đầu
encoded_columns = encoder.get_feature_names_out(columns_to_encode)
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)
data = pd.concat([data, encoded_df], axis=1)

# Xóa các cột ban đầu
data = data.drop(columns=columns_to_encode)

# Lưu kết quả vào file mới
data.to_csv('Train_Samsung_imputed_onehot.csv', index=False)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Đọc dữ liệu huấn luyện từ file CSV
train_data = pd.read_csv('Train_Samsung_imputed_onehot_with_class.csv')
X_train = train_data.drop(columns=['class'])
y_train = train_data['class']

# Đọc dữ liệu kiểm tra từ file CSV
test_data = pd.read_csv('Test_samsung_noclass_imputed_onehot.csv')
X_test = test_data

# Khởi tạo mô hình KNN

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

# Dự đoán nhãn cho dữ liệu kiểm tra
y_pred = knn.predict(X_test)

# Thêm nhãn vào DataFrame ban đầu
test_data['class'] = y_pred

# Lưu kết quả vào file mới
test_data.to_csv('Test_samsung_with_class.csv', index=False)
```

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score

# Đọc dữ liệu từ file Train_Samsung_imputed.csv
train_df = pd.read_csv('Train_Samsung_imputed.csv')

# Bỏ các cột không mong muốn
columns_to_drop = ['X1', 'X2', 'X3', 'X4', 'X5', 'X11']
train_df = train_df.drop(columns_to_drop, axis=1)

# Chuẩn bị dữ liệu cho việc huấn luyện mô hình Naive Bayes
X = train_df.drop('Class', axis=1)
y = train_df['Class']

# Huấn luyện mô hình Naive Bayes
gnb = GaussianNB()

# Thực hiện kiểm định chéo 5-fold
scores = cross_val_score(gnb, X, y, cv=5, scoring='accuracy')

# In kết quả
print(f'Tỷ lệ chính xác: {scores.mean():.2f} (+/- {scores.std() * 2:.2f})')
```
