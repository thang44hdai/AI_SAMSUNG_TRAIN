```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AB_NYC_2019 (1).csv")

neighbourhood = data['neighbourhood_group']
group = neighbourhood.value_counts()
plt.bar(group.index, group.values, color = "#f25eb7")

# Nằm ngang
#plt.barh(data['day'], data['tip'])
plt.title("Neighborhood Group")

plt.xlabel('Neighborhood Group')
plt.ylabel('Count')

plt.show()
```

```python
res = data["neighbourhood"]
tmp = res.value_counts()
plt.hist(tmp, bins=80)

plt.xlabel("Số lượng")
plt.ylabel("Tần suất")
plt.title("Histogram")

plt.show()
```
