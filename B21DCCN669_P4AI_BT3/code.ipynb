## KNN (K láng giềng gần)

```python
import pandas as pd
import numpy as np

# Đọc dữ liệu từ file csv
df = pd.read_csv('P4AI_BT1.csv')

# Tính giá trị trung bình cho mỗi cột
mean_sepal_length = df['sepal.length'].mean()
mean_sepal_width = df['sepal.width'].mean()
mean_petal_length = df['petal.length'].mean()
mean_petal_width = df['petal.width'].mean()

# Điền giá trị trung bình vào các ô trống
df['sepal.length'] = df['sepal.length'].replace(np.nan, mean_sepal_length)
df['sepal.width'] = df['sepal.width'].replace(np.nan, mean_sepal_width)
df['petal.length'] = df['petal.length'].replace(np.nan, mean_petal_length)
df['petal.width'] = df['petal.width'].replace(np.nan, mean_petal_width)

# Lưu kết quả vào file mới
df.to_csv('filled_data.csv', index=False)
pd.set_option('display.max_rows', None)
df = pd.read_csv('filled_data.csv')

# In ra nội dung của file csv
display(df)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
# Đọc dữ liệu từ file csv
df = pd.read_csv('filled_data.csv')

# Tách tập dữ liệu thành 2 phần: có nhãn và chưa có nhãn
labeled_df = df[df['variety'].notnull()]
unlabeled_df = df[df['variety'].isnull()]

# Huấn luyện mô hình k-NN
X_train = labeled_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y_train = labeled_df['variety']
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Dự đoán nhãn cho phần dữ liệu chưa có nhãn
X_test = unlabeled_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y_pred = knn.predict(X_test)
unlabeled_df['variety'] = y_pred

# Gộp 2 phần dữ liệu lại với nhau
df = pd.concat([labeled_df, unlabeled_df])

# Lưu kết quả vào file mới
df.to_csv('KNN.csv', index=False)
pd.set_option('display.max_rows', None)
df = pd.read_csv('KNN.csv')

# In ra nội dung của file csv
display(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
df = pd.read_csv('KNN.csv')

X = labeled_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y = labeled_df['variety']

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Huấn luyện mô hình k-NN trên tập huấn luyện
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Dự đoán nhãn cho các hàng trong tập kiểm tra
y_pred = knn.predict(X_test)

# Tính toán tỉ lệ chính xác
accuracy = accuracy_score(y_test, y_pred)
accuracy = round(accuracy, 3)
print(f'Tỉ lệ chính xác: {accuracy}')
```

```python
from sklearn.metrics import f1_score

# Tính toán F1 score
f1 = f1_score(y_test, y_pred, average='weighted')
f1 = round(f1, 2)
print(f'F1 score: {f1}')
```
