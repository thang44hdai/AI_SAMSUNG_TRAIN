```python
import pandas as pd

data = pd.read_csv("spam.csv", encoding='latin-1')

# Xóa cột vô nghĩa
data = data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)

# Lưu dữ liệu train
train_emails = data['v2'].values
train_labels = data['v1'].values

# Xây dựng và huấn luyện mô hình
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(CountVectorizer(), MultinomialNB())
model.fit(train_emails, train_labels)

# Dự đoán nhãn
predicted_label = model.predict(data['v2'])

# Thêm cột dự đoán vào DataFrame
data['predicted_label'] = predicted_label

# In dữ liệu sau khi thêm cột dự đoán
display(data)
```
