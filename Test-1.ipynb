```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('Train_samsung.csv')
data
```

```python
data['X1'].fillna(data['X1'].mode()[0],inplace = True)
data['X2'].fillna(data['X2'].mode()[0],inplace = True)
data['X3'].fillna(data['X3'].mode()[0],inplace = True)
data['X4'].fillna(data['X4'].mode()[0],inplace = True)
data['X5'].fillna(data['X5'].mode()[0],inplace = True)
data['X6'].fillna(round(data['X6'].mean()),inplace = True)
data['X7'].fillna(round(data['X7'].mean()),inplace = True)
data['X8'].fillna(round(data['X8'].mean()),inplace = True)
data['X9'].fillna(round(data['X9'].mean()),inplace = True)
data['X10'].fillna(data['X10'].mode()[0],inplace = True)
data['X11'].fillna(data['X11'].mode()[0],inplace = True)
data
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
label = LabelEncoder()
data2 = data.copy()
train = data.copy()
for c in  data2.columns:
    if(data2[c].dtype=='object'):
        train[c]=label.fit_transform(data2[c])
    else:
        train[c]=data2[c]
train
```

```python
X = train.iloc[:,:-1]
y = train.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y ,random_state=104, test_size=0.2, shuffle=True)
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score


clf = SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("SVM accuracy: ", accuracy_score(y_test, y_pred))
print("SVM f1 score: ", f1_score(y_test, y_pred, average='weighted'))
print("SVM confusion matrix: ")
print(confusion_matrix(y_test, y_pred))
```

```python
test_ss = pd.read_csv('Test_samsung_noclass.csv')
test_ss
```

```python
test_ss.info()
```

```python
test_ss['X1'].fillna(test_ss['X1'].mode()[0],inplace = True)
test_ss['X2'].fillna(test_ss['X2'].mode()[0],inplace = True)
test_ss['X3'].fillna(test_ss['X3'].mode()[0],inplace = True)
test_ss['X4'].fillna(test_ss['X4'].mode()[0],inplace = True)
test_ss['X5'].fillna(test_ss['X5'].mode()[0],inplace = True)
test_ss['X6'].fillna(round(test_ss['X6'].mean()),inplace = True)
test_ss['X7'].fillna(round(test_ss['X7'].mean()),inplace = True)
test_ss['X8'].fillna(round(test_ss['X8'].mean()),inplace = True)
test_ss['X9'].fillna(round(test_ss['X9'].mean()),inplace = True)
test_ss['X10'].fillna(test_ss['X10'].mode()[0],inplace = True)
test_ss['X11'].fillna(test_ss['X11'].mode()[0],inplace = True)
test_ss
```

```python
df = test_ss.copy()
train_ss = test_ss.copy()
for x in  df.columns:
    if(df[x].dtype=='object'):
        train_ss[x]=label.fit_transform(df[x])
    else:
        train_ss[x]=df[x]
        
train_ss
```

```python
y_pred_ss = clf.predict(train_ss)
result = pd.DataFrame(y_pred_ss, columns = ['class'])
horizontal_concat = pd.concat([train_ss, result], axis=1)
horizontal_concat
```

```python
horizontal_concat.to_csv('final_file.csv')
```
